# Task 1.3 — What the Paper Claims to Improve

**Paper:** Nyström Method vs Random Fourier Features: A Theoretical and Empirical Comparison (NIPS 2012)

---

## Main Baseline

The main baseline method that the paper compares against is **Random Fourier Features (RFF)**, introduced by Rahimi and Recht (2007, 2008). The paper treats RFF as the primary prior method and positions Nyström as the alternative whose theoretical advantages it seeks to characterise and justify.

---

## Limitation of the Baseline (RFF)

RFF's core limitation, as identified by the paper, is that it uses **data-independent random projections** — the frequency vectors ω are sampled from the kernel's spectral distribution without looking at the training data at all. This means RFF cannot adapt to the structure of the data. Specifically, if the training data lies on a low-dimensional manifold or the kernel matrix has fast-decaying eigenvalues, RFF wastes its budget D on directions that are spectrally irrelevant — directions where the kernel matrix has negligible energy. The paper shows that RFF's approximation error scales as O(D^{-1/2}) regardless of the data's spectral structure, which is suboptimal when better structure exists.

---

## How the Paper Overcomes This Limitation

The Nyström method overcomes this by using **data-adaptive landmark sampling**: it selects basis points directly from the training data, which means the feature map it constructs naturally aligns with the directions of highest spectral energy in the actual kernel matrix. The paper proves that Nyström's error is bounded by the tail of the kernel matrix's eigenspectrum — which is small when eigenvalues decay quickly — giving it a fundamentally tighter bound than RFF in structured settings.

---

## Condition Under Which the Paper's Method Would NOT Outperform the Baseline

Based on my reading of Theorem 2 and the experimental analysis in Section 4, Nyström would **not** outperform RFF in a scenario where the kernel matrix has a near-flat eigenspectrum — that is, where all eigenvalues are approximately equal. In this case, Nyström's error bound (which depends on the tail eigenvalue sum) does not decay faster than RFF's O(D^{-1/2}) bound, eliminating the theoretical advantage. Practically, this occurs when the training data is uniformly distributed in feature space with no dominant directions — for example, white noise data or data arising from a uniform distribution over a high-dimensional cube. In such a setting, Nyström's data-adaptive landmark selection provides no useful information that RFF's random projections do not already capture, and RFF may even outperform Nyström because it avoids the eigendecomposition step that can amplify noise when K_{mm} is nearly singular.

Additionally, when the budget D is very large (approaching n, the training set size), both methods converge to the exact kernel and the difference between them vanishes — making the choice irrelevant in high-budget settings. The paper acknowledges this implicitly in the convergence analysis.
