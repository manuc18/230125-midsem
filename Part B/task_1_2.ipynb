# Task 1.2 — Key Assumptions

**Paper:** Nyström Method vs Random Fourier Features: A Theoretical and Empirical Comparison (NIPS 2012)

---

## Assumption 1

**Assumption:** The kernel function is shift-invariant, i.e., k(x, y) = k(x − y), and satisfies Bochner's theorem so it has a well-defined spectral distribution p(ω).

**Why the method needs it:** RFF is grounded entirely in the Fourier representation of shift-invariant kernels. If the kernel is not shift-invariant (e.g., polynomial kernel k(x,y) = (xᵀy + c)^d), then no spectral distribution p(ω) exists, and the random frequency sampling procedure in Step 1 of RFF has no theoretical justification. Nyström also benefits from this property since its approximation quality is analysed in terms of the kernel matrix's eigenspectrum, which has cleaner structure for shift-invariant kernels.

**Violation scenario:** If a practitioner uses a polynomial or sigmoid kernel on text data (where features are discrete counts), the shift-invariance assumption fails and the RFF approximation becomes invalid. The paper's theoretical guarantees do not extend to such kernels.

**Paper reference:** Section 2, Preliminaries — the paper opens by restricting all analysis to shift-invariant kernels and invokes Bochner's theorem explicitly before deriving the RFF feature map.

---

## Assumption 2

**Assumption:** The eigenvalues of the kernel matrix K decay rapidly (the eigenspectrum has a "sharp drop-off"), so the matrix has low effective rank.

**Why the method needs it:** The paper's central theoretical result (Theorem 1) shows that Nyström's approximation error is bounded by Σ_{i>m} λᵢ — the sum of eigenvalues beyond rank m. This bound is only tight when eigenvalues decay fast, meaning a small m suffices to capture most spectral energy. If the spectrum is flat (all λᵢ equal), Nyström's bound degrades and its advantage over RFF disappears, as proved in Theorem 2 comparison.

**Violation scenario:** Data generated from a uniform distribution over a high-dimensional hypercube produces a nearly flat kernel matrix eigenspectrum. In this setting, no small subset of m landmarks can represent the kernel matrix well, and the Nyström method provides no advantage over RFF.

**Paper reference:** Section 3 (Theoretical Analysis), Theorems 1 and 2 — the comparison between bounds explicitly shows the eigenvalue decay rate as the deciding factor. Section 4 (Experiments) also validates this on datasets with different spectral properties.

---

## Assumption 3

**Assumption:** The landmark points sampled for the Nyström approximation are representative of the training data distribution — i.e., they are sampled uniformly or via clustering from the actual training set.

**Why the method needs it:** Nyström's approximation quality depends on how well the m landmark points span the space of the training data in kernel space. The error bound in Theorem 1 is derived under the assumption that landmarks are sampled uniformly from the training set. If landmarks are chosen poorly (e.g., from a different distribution, or all from one cluster), the sub-matrix K_{mm} does not accurately reflect the full kernel matrix's spectral structure, and the bound no longer holds.

**Violation scenario:** In a multi-class dataset with highly imbalanced classes, uniform random sampling of landmarks may draw almost all points from the majority class, leaving minority-class structure unrepresented. The Nyström approximation would then be poor for minority-class samples, causing degraded SVM performance on those classes.

**Paper reference:** Section 2.2 (Nyström Method description) and Section 4 (Experiments) — the paper uses uniform subsampling of training points as landmarks throughout. The theoretical analysis in Section 3 implicitly assumes this sampling strategy by conditioning on landmarks drawn i.i.d. from the data distribution.
